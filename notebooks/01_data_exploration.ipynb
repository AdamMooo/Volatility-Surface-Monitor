```python
import sys
sys.path.insert(0, '../src')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

from data.fetcher import OptionChainFetcher, RiskFreeRateFetcher
from data.cleaner import clean_option_chain
```

```python
# Fetch SPY option chain
fetcher = OptionChainFetcher('SPY')
chain = fetcher.fetch_current_chain()

print(f"Raw chain shape: {chain.shape}")
chain.head(10)
```

```python
# Clean the chain
clean_chain = clean_option_chain(chain)

print(f"Cleaned chain shape: {clean_chain.shape}")
clean_chain.head(10)
```

```python
# Visualize moneyness distribution
fig = px.histogram(clean_chain, x='moneyness', nbins=50, 
                   title='Moneyness Distribution')
fig.show()
```

```python
# IV by strike
calls = clean_chain[clean_chain['option_type'] == 'call']
fig = px.scatter(calls, x='moneyness', y='impliedVolatility', 
                 color='days_to_expiry',
                 title='IV vs Moneyness (Calls)')
fig.show()
```
