```python
import sys
sys.path.insert(0, '../src')

import pandas as pd
import numpy as np
import plotly.graph_objects as go

from data.fetcher import OptionChainFetcher
from data.cleaner import clean_option_chain
from models.surface import IVSurface
from visualization.surface_plots import plot_iv_surface, plot_smile_comparison
```

```python
# Fetch and clean data
fetcher = OptionChainFetcher('SPY')
chain = fetcher.fetch_current_chain()
clean_chain = clean_option_chain(chain)

print(f"Clean chain: {len(clean_chain)} options")
```

```python
# Build IV surface
surface = IVSurface(clean_chain, method='rbf')

# Get ATM vol
atm_30d = surface.get_atm_vol(30)
atm_60d = surface.get_atm_vol(60)

print(f"30-day ATM Vol: {atm_30d*100:.2f}%")
print(f"60-day ATM Vol: {atm_60d*100:.2f}%")
```

```python
# Plot 3D surface
fig = plot_iv_surface(surface)
fig.show()
```

```python
# Plot volatility smiles
fig = plot_smile_comparison(surface, expiries=[14, 30, 60, 90])
fig.show()
```

```python
# Term structure
term_struct = surface.get_term_structure()

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=list(term_struct.keys()),
    y=list(term_struct.values()),
    mode='lines+markers',
    name='ATM Term Structure'
))
fig.update_layout(
    title='ATM Volatility Term Structure',
    xaxis_title='Days to Expiry',
    yaxis_title='Implied Volatility'
)
fig.show()
```
